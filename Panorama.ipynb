````python
import cv2 as cv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
import math
````

```
Steps for creating panorama using open cv
```

````python
#Reaing Images
images=glob.glob(f"boat{os.sep}*")
images.sort()
imagefiles=[]
for filename in images:
  im=cv.imread(filename,1)
  im=cv.cvtColor(im,cv.COLOR_BGR2RGB)
  imagefiles.append(im)
num=len(imagefiles)
print("No of Images:",num)
````

````python
from matplotlib import image
#Display Images
plt.figure(figsize=[30,10])
num_cols=3
num_row=math.ceil(num/num_cols)
for i in range(0,num):
  plt.subplot(num_row,num_cols,i+1)
  plt.axis("off")
  plt.imshow(imagefiles[i])
````

```
Use sticher class
```

````python
#intiate a sticher class
stitcher=cv.Stitcher_create()
status,im=stitcher.stitch(imagefiles)
if status==0:
  plt.figure(figsize=[30,10])
  plt.imshow(im)
````
