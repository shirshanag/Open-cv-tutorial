```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import cv2 as cv
from IPython.display import Image
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
```

```python
img=cv.imread("/content/Apollo_11_Launch.jpg",1)
im=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(im)
```

```python
#Add line in the picture
img_copy=img.copy()
cv.line(img_copy,(200,100),(400,100),(0,255,255),thickness=5,lineType=cv.LINE_AA);
#(200,100)--.starting point
#(400,100)-->ending point

plt.imshow(img_copy[:,:,::-1])
```

```python
#Adding circle in images
img1_Copy=img.copy()
cv.circle(img1_Copy,(900,500),100,(0,255,255),thickness=-5,lineType=cv.LINE_AA)
#(900,500)-->centre of the circle
#100 ->>radius of the circle
plt.imshow(img1_Copy[:,:,::-1])
```

```python
#Adding Rectangle
img_rect=img.copy()
cv.rectangle(img_rect,(500,100),(700,600),(255,0,255),thickness=5,lineType=cv.LINE_AA)
## Here (500,100) is the top left vertex of the the rectangle,(700,600) is the bottom right vertex of the rectangle

plt.imshow(img_rect[:,:,::-1])
```

```python
#Adding text in the image
img3_copy=img.copy()
text="Apollo-11-Launch"
font_scale=2
font_face=cv.FONT_HERSHEY_SIMPLEX
font_color=(0,255,0)
font_Thickness=2
cv.putText(img3_copy,text,(200,700),fontFace=font_face,fontScale=font_scale,color=font_color,thickness=font_Thickness,lineType=cv.LINE_AA)
#text-->is the text we want to write in the pic
#fontface--> the style of text
#font scale-->Font scale factor that is multiplied by the font-specific base size.
#font color-->font color of the text
#font thickness-->font thickness of the text
plt.imshow(img3_copy[:,:,::-1])
```
